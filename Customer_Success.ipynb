## Issue Resolution Time

```python
# Agregar columna con tiempo de resolución
data['resolve_time_hrs'] = round(((data.Resolved - data.Created).dt.total_seconds() / 60 / 60), 2)
```

## Gráfica - Tiempo promedio de solución de tickets por semana (hr)

### Preparar data

```python
# Filtrar solo por tickets que ya fueron completados
data_resolucion_promedio = data_f[data_f.State == 'Resolved'].copy()
```

```python
metricas_por_dia = pd.DataFrame({'dia': pd.date_range(start='2022-01-03', end=date.today().strftime('%Y-%m-%d')),
                           })
metricas_por_dia.head()
```

```python
# Corregir tipo de data en las columnas
data_resolucion_promedio['fecha_creacion'] = pd.to_datetime(data_resolucion_promedio['fecha_creacion'], format = "%Y-%m-%d %H:%M:%S")
data_resolucion_promedio['resolve_time_hrs'] = data_resolucion_promedio['resolve_time_hrs'].astype('float')
```

```python
data_resolucion_promedio
```

```python
# Calcular tiempo de resolución promedio de tickets por día
tiempo_resolucion_por_ticket_por_dia = data_resolucion_promedio.resample('D', on='fecha_creacion')['resolve_time_hrs'].mean()

# Mezclar con el DataFrame general
metricas_por_dia = metricas_por_dia.merge(tiempo_resolucion_por_ticket_por_dia, how='left', left_on='dia', right_index=True).rename(columns={'resolve_time_hrs': 'avg_resolve_time_hrs'}).fillna(0)
metricas_por_dia.tail()
```

Añadir ahora los usuarios registrados en cada día

```python
# Calcular tiempo de resolución promedio de tickets por día
registrados_por_dia = df_tau.resample('D', on='signup_cohort_date')['id'].count()

# Mezclar con el DataFrame general
metricas_por_dia = metricas_por_dia.merge(registrados_por_dia, how='left', left_on='dia', right_index=True).rename(columns={'id': 'registrados'}).fillna(0)
metricas_por_dia.tail()
```
